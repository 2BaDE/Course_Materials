## 3.1 While statement 

- The **while statement** repeats a body of code as long as a condition is True. 
- If the condition is True, the body of the statement will execute.

```Python
counter = 1

while counter <= 5:
    print("Hello, world")
    counter = counter + 1
```

## 3.2 For statement 

- The **for statement** can be used to iterate over the members of a collection, so long as the collection is a sequence. 
- The for statement works for any collections that is a seuquence (lists, tuples, and strings). 
- A common use of the for statement : To implement definite iteration over a range of values. 

```Python
wordlist = ['cat','dog','rabbit']
letterlist = [ ]
for aword in wordlist:
    for aletter in aword:
        letterlist.append(aletter)
print(letterlist)
```

## 3.4 List Comprehension

- A list comprehension allows us to easily create a list based on some processing or selection criteria. 

```Python
# Compute and add to the list 
sqlist=[x*x for x in range(1,11)]

# Selecction criteria of list comprehension
sqlist=[x*x for x in range(1,11) if x%2 != 0]
```

# 4. Exception Handling 

## 4.1 Kinds of Errors 

- **Syntax error** 
    - Syntax error means that the programmer has made a mistake in the structure of a statement or expression. 
    - Python interpreter found it cannot complete the processing of this instruction since it does not conform the rules of the language. 
- **Logical error** 
    - Denotes a situation where the program executes but gives the wrong result. 
    - Logical error leads to a runtime error that cause the program to terminate. 
    - These types of runtime errors are typically called **exceptions** 
    
## 4.2 try statement 

Programmers can create their own exceptions if they detect a situation in the program execution that warrants it. When an excpetion occurs, we say that it has been "raised'. We can handle the exception that has been raised by using a try statement. 

We can handl this exception by writing codes within a try block. A corresponding except block catches the exception and prints a message back to the user in the event that an exception occurs. 

```Python
try:
    print(math.sqrt(anumber))
except:
    print("Bad Value for square root")
    print("Using absolute value instead")
    print(math.sqrt(abs(anumber)))
```

## 4.3 raise statement 

It is also possible for a programmer to cause a runtime exception by using the raise statement. 

```Python
if anumber < 0:
    raise RuntimeError("You can&#39;t use a negative number")
else:
    print(math.sqrt(anumber))
```

## 6.1 Fraction Class 

### Defining Class

- All classes should provide **constructor**. 
- The constructors defines the way in which data objects are created.

- **self** is a special parameter that will always be used as a reference ack to the object itself.

### Implementation of Fraction Class 

- The operations for the Fraction type : add, subtract, multiply, divide
- Can show fractions using the standard "slash" form. 
- Fraction needs two pieces of data, the numerator and the denominator. 

```Python
class Fraction: 
    
    def __init__(self, top, bottom):
        self.num = top
        self.den = bottom 
        
myf = Fraction(3, 5) 
```

### Apply Standard methods to Fraction Class

**1. The problem of printing instance of Class** 

When we print the instance of class using print function, our return will be same as below.

```Python
myf = Fraction(3,5) 
print(myf) 
# <__main__.Fraction instance at 0x409b1acc>
```

This is because, the print function requires that the object convert itself into a string. The only choice instance has is to show the actual reference that is stored in the variable. 

**2. Solving problem** 

There are two ways we can solve this problem. One is defining a personal method to show, and the other is using a set of standard methods. 

**3. Standard methods**

In Python, all classes have a set of standard methods that are provided but may not work properly. **\_\_str\_\_** is the method to convert an object into a string.

However, the default implementation of standard methods is to return the instance address string. So, to do better implementation, we need to do **implementation overrides** which means redefining the method's behavior. 

The method will build a string representation by converting each piece of internal state data to a string and then placing a / character in between the strings using string concatenation. 

We can apply other standard mathods to implement arithmetic operations. 

```Python
class Fraction: 

    def __init__(self, top, bottom):
        self.num = top
        self.den = bottom 
        
    def __str__(self): 
        return str(self.num) + "/" + str(self.den)
    
myf = Fraction(1, 4)
print(myf)
# 1 / 4
```

**4. Overrides \_\_add\_\_ of Fraction class** 

We can use "+" operator to fraction object by overriding \_\_add\_\_ methods. 

```Python
class Fraction: 

    def __init__(self, top, bottom):
        self.num = top
        self.den = bottom 
        
    def __str__(self): 
        return str(self.num) + "/" + str(self.den)
    
    def __add__(self, other_fraction): 
        newnum = self.num*other_fraction.den + self.den*other_fraction.num
        newden = self.den*other_fraction.den 
        return Fraction(newnum, newden) 
    
f1 = Fraction(1, 4)
f2 = Fraction(1, 2) 
# 6 / 8
```

**4-1. Apply lowest terms representation** 

In previous example, the method works as we desire, but one thing could be better. Note that 6/8 is the correct result but that is not in the "lowest terms" representation. We can then divide the numerator and the denominator by the GCD. 

```Python
class Fraction: 
    
    def __init__(self, top, botton): 
        self.num = top
        self.den = botton 
    
    def __str__(self): 
        return str(self.num) + "/" + str(self.den) 
    
    def _gcd(self, m, n): 
        while m % n != 0:
            oldm = m
            oldn = n 
            
            m = oldn
            n = oldm % oldn
            
        return n 
    
    def __add__(self, other_fraction): 
        newnum = self.num*other_fraction.den + self.den*other_fraction.num
        newden = self.den*other_fraction.den 
        gcd = self._gcd(newnum, newden) 
        return Fraction(newnum//gcd, newden//gcd)
    
f1 = Fraction(1, 4)
f2 = Fraction(1, 2) 
# 3 / 4    
```

**5. Overrides \_\_eq\_\_ standard method** 

The \_\_eq\_\_ method compares two objects and returns True if their values are the same, False otherwise. In the Fraction class, we can implement the \_\_eq\_\_ method by again putting the two fractions in common term and then comparing the numerators. 

```Python
def __eq__(self, other): 
    firstnum = self.num * other.den 
    secondnum = other.num * self.den 
    return firstnum == secondnum
```

## 6.2 Logic Gates and Circuits Class 

### Class Inheritance

- **Inheritance** is the ability for one class to be related to another class. 
- Python chlid classes can inherit characteristic data and behavior from a parent class. 

![](Img/Injeritance.png)

- Above structure is called **inheritance hierarchy**. 
- This implies that lists inherit important characteristics from sequences, namely the ordering of the underlying data and the operations such as concatenation, repetition, and indexing. 
- The children all gain from their parent but distinguish themselves by adding additional characteristics. 
- By organizing classes in this hierrachical fashion, object-oriented programming languages allow previously written code to be extended to meet the needs of a new situation. 

### Logic Gate 

![](Img/LogicGates.png)

```Python
class LogicGate: 
    
    def __init__(self, n): 
        self.label = n
        self.output = None 
        
    def getLabel(self): 
        return self.label 
    
    def getOutput(self): 
        self.output = self.performGateLogic()
        return self.output
```

The powerful idea in OOP is that we can write a method that will use code that does not exist yet. 

### Binary/UnaryGate

```Python
class BinaryGate(LogicGate): 
    
    def __init__(self, n): 
        LogicGate.__init__(self, n)
        
        self.pinA = None
        self.pinB = None 
        
    def getPinA(self): 
        return int(input(f"Enter Pin A input for gate {self.getLabel()} -->"))
    
    def getPinB(self): 
        return int(input(f"Enter Pin B input for gate {self.getLabel()} -->")) 
    
class UnaryGate(LogicGate): 
    
    def __init__(self, n): 
        LogicGate.__init__(self, n)
        
        self.pin = None 
        
    def getPin(self): 
        return int(input(f"Enter Pin input for gate {self.getlabel()}"))
```

The constructors in both of these classes start with an explicit call to the constructor of the parent class using the parent's \_\_init\_\_ method. In this case, that means the label for the gate. Then the constructor then goes on to add the input lines (pinA and pinB). 


### AND/OR/NOTGate

```Python
class AndGate(BinaryGate): 
    
    def __init__(self, n): 
        super(AndGate, self).__init__(n)
        
    def performGateLogic(self): 
        
        a = self.getpinA()
        b = self.getpinB() 
        if a == 1 and b == 1: 
            return 1 
        else : 
            return 0
```

The only thing And/Or/NotGate needs to add is the specific behavior that performs the boolean operation that was described. This is the place where we can provide the perfromGateLogic method. 

Note that the AndGate class does not provide any new data since it inherits two input liens, one output line, and a label. 

### Connector Class 

With the Connector class, Connetctors will have instances of the LogicGate class wihtin them but are not part of the hierarchy. When designing classes, it is very important to distinguish between those that have the IS-A relationship (which requires inheritance) and those that have HSA-A relationships (with no inheritance). 

```Python
class Connector: 
    
    def __init__(self, fgate, tgate): 
        self.fromgate = fgate
        self.togate = tgate
        
        tgate.setNextPin(self)
        
    def getFrom(self): 
        return self.fromgate
    
    def getTo(self): 
        return self.togate 
```
